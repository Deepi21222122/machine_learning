# HANDLING MISSING VALUES

```python
data.isnull().sum() #checking out the missing values
```

## DATA CLEANING

```python
data.isnull().sum()/data.shape[0]*100
```

```python
# data[col].isnull().sum()/data.shape[0] * 100 < 70)
#Commented line to check if that column's null percentage is below 70

cols = [col for col in data.columns if (data[col].isnull().sum()/data.shape[0] * 100 < 70)]
data_set = data[cols]

print ('Legitimate columns after dropping null columns: %s' % data_set.shape[1])
```

```python
data_set.isnull().sum()
```

```python
data_set.head()
```

```python

#Looks like some columns needs to be converted to numeric field

data_set['Snowfall'] = pd.to_numeric(data_set['Snowfall'], errors='coerce')
data_set['SNF'] = pd.to_numeric(data_set['SNF'], errors='coerce')
data_set['PRCP'] = pd.to_numeric(data_set['PRCP'], errors='coerce')
data_set['Precip'] = pd.to_numeric(data_set['Precip'], errors='coerce')

data_set['Date'] = pd.to_datetime(data_set['Date'])
```

```python
#Fill remaining null values. FOr the moment lts perform ffill

data_set.fillna(method='ffill', inplace=True)
data_set.fillna(method='bfill', inplace=True)

data_set.isnull().sum()
#Well no more NaN and null values to worry about
```

```python
X = data_set.iloc[:, 4].values
y = data_set.iloc[:, 3].values
```

```python
X
```

```python
y
```

```python
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
X.shape
```

```python
y.shape
```

# CREATING TRAIN AND TEST LABELS

```python
from sklearn.model_selection import train_test_split     # separating train set  and test set values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
plt.rcParams['figure.figsize']=[12,12] #Scattered diagram
plt.scatter(X,y,s=65)
plt.xlabel('Min Temp')
plt.ylabel('Max Temp')
plt.title('Min Vs Max Temp')
plt.show()
```

# 1.Prediction of data using linear regression model

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr=Lr.fit(X_train, y_train)
```

```python
ypr= Lr.predict(X_test)
```

```python
plt.scatter(X_train,y_train,c='green',label='train-part')
plt.scatter(X_test,y_test,c='blue',label='test-part')
plt.plot(X_test,Lr.predict(X_test),c='red')
plt.xlabel("Min Temp")
plt.ylabel("Max Temp")
plt.legend()
plt.title("Min Temp Vs Max Temp for both train and test part")
plt.show()
```

```python
a =Lr.coef_        
c =Lr.intercept_
print("y = ",c ," + ",a," x  ")  #intercept and coefficient of linear model 
```

```python
from sklearn.metrics import r2_score     # whenever r2_score is very close to 1 then prediction is fine.
r2 =r2_score(y_test,ypr)
print("R_square value of linear regression model is ",r2)
R_square={}
R_square["Linear Regression R_Square Score:"]=r2
```

###  Predicting the maximum temperature for given the minimum temperature

```python
min_temp =17
y_pr = Lr.predict(min_temp).flatten()[0]
print("max temp = ",y_pr,"for given min temp",min_temp)
```
